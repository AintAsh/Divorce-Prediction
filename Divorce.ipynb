```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
col = ['education_level','employment_status','religious_compatibility','conflict_resolution_style','marriage_type']
for i in col:
    df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

```python
X = df.drop('divorced',axis=1)
y = df['divorced']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
def evaluate(model, X_train, y_train, X_test, y_test):

    # Predictions
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    # Probabilities (needed for ROC-AUC)
    train_prob = model.predict_proba(X_train)[:, 1]
    test_prob = model.predict_proba(X_test)[:, 1]

    print(f"Evaluation for {model}")
    print("-" * 50)

    # Accuracy
    print(f"Training Accuracy for {model} is {accuracy_score(y_train, train_pred)}")
    print(f"Testing Accuracy for {model} is {accuracy_score(y_test, test_pred)}")

    print("\nConfusion Matrix (Test):")
    print(confusion_matrix(y_test, test_pred))

    print("\nClassification Report (Test):")
    print(classification_report(y_test, test_pred))

    # ROC-AUC
    print(f"Training ROC-AUC for {model} is {roc_auc_score(y_train, train_prob)}")
    print(f"Testing ROC-AUC for {model} is {roc_auc_score(y_test, test_prob)}")
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
evaluate(lr,X_train,y_train,X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(X_train,y_train)
```

```python
evaluate(kn,X_train,y_train,X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42
)

dt.fit(X_train,y_train)
```

```python
evaluate(dt,X_train,y_train,X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
    n_estimators=400,
    max_depth=6,
    min_samples_split=30,
    min_samples_leaf=15,
    max_features="sqrt",
    class_weight="balanced_subsample",
    random_state=42

    )
rf.fit(X_train,y_train)
```

```python
evaluate(rf,X_train,y_train,X_test,y_test)
```

```python
import pickle

with open("divorce_model.pkl", "wb") as fs :
    pickle.dump(rf,fs)
```
